In this file, we will show how to convert the DICOM images into the standard image to make the Classifier run more efficient in terms of memory. You need to download the entire dataset from CBIS-DDSM images before run these blocks of codes.

```python
#Library needed for converting the image format
import os
import numpy as np
import pydicom
from PIL import Image
```

Below is the function to get the pathfile name before converting the image. The parameter for this function is the directory/folder where the image file located. For our case, the directory/folder name is CBIS-DDSM.

```python
def get_name(directory):

    names = []

    for root, dirs, files in os.walk(directory):
        for filename in files:
            names.append(os.path.join(root, filename))

    return names
#Initiate the variable to get the pathfile for each images
names = get_name('CBIS-DDSM')
#Showing how does it looks like
names.head(5)
```

Below is the function for converting the image. The parameter for this function is the pathfile that we have generated from previous code. In our case, we named it 'names'.

```python
def convert_image(pathfile):
    im = pydicom.dcmread(pathfile)
    im = im.pixel_array.astype(float)

    rescaled_image = (np.maximum(im,0)/im.max())*255 # float pixels
    final_image = np.uint8(rescaled_image) # integers pixels

    final_image = Image.fromarray(final_image)
    return final_image
```

```python
for name in names:
    image = convert_image(name)
    image.save(name+'.png')
```

Now, you can check the png format images in the directory that have been generated by running all the codes above. Apparently, the memory for one image significantly reduce (e.g. from 26 Mb to 2 Mb) which could help the classifier run faster. Though the images have been converted, there are no changes in pixel size for each image.
